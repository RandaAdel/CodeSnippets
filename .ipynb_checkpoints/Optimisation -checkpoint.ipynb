```python
import numpy as np
from scipy.optimize import minimize
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('advertising.csv', parse_dates=['Date'])
df.head()
```

```python
# create a copy of the original dataframe
df_copy = df.copy()

# drop the Date and sales columns from the copied dataframe
X = df_copy.drop(['Date', 'Sales'], axis=1)
y = df_copy['Sales']

# Split the data into training and testing sets
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=42)
```

```python
# Input data - columns represent TV, Radio, and Newspaper spending
# X = np.array([[200, 50, 100],
#               [150, 25, 75],
#               [300, 75, 50],
#               [175, 35, 90]])

# # Output data - represents total sales
# y = np.array([1500, 1000, 2000, 1200])

X = train_X
y = train_y

# Define the objective function to minimize
def objective(weights, X, y):
    # Compute predicted sales using the linear regression model
    predicted_sales = sm.OLS(y, X).fit().predict(sm.add_constant(weights.reshape(1, -1)))
    # Return negative of predicted sales to maximize sales
    return -predicted_sales

# Define the constraints
def constraint(weights):
    # Ensure total spending does not exceed 100,000
    return 1000 - np.sum(weights)

# Define the bounds for each weight (spending)
bounds = [(0, 1000)] * X.shape[1]

# Define the initial guess (equal spending for all channels)
x0 = np.ones(X.shape[1]) * (1000 / X.shape[1])

# Solve the optimization problem
result = minimize(objective, x0, args=(X, y), method='SLSQP', bounds=bounds, constraints={'type': 'ineq', 'fun': constraint})

# Print the optimized weights (spending) for each channel
print("Optimized weights (TV, Radio, Newspaper): ", result.x)

# Compute the predicted sales using the optimized weights
sales_model = sm.OLS(y, X).fit()
predicted_sales = sales_model.predict(sm.add_constant(result.x.reshape(1, -1)))
print("Predicted sales with optimized weights: ", predicted_sales)
```
